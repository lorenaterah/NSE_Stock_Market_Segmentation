```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('../Data/NSE_data_all_stocks_2021_upto_31dec2021.csv')
df.head()
```

```python
sectors_2021 = pd.read_csv('../Data/NSE_data_stock_market_sectors_as_at_31dec2021.csv')
sectors_2021.head()
```

```python
# to attach sectors to df using the code
df = df.merge(sectors_2021[['CODE', 'SECTOR']], on='CODE', how='left')
df.head()
```

```python
df1 = pd.read_csv('../Data/NSE_data_all_stocks_2022.csv')
df1.head()
```

```python
sectors_2022 = pd.read_csv('../Data/NSE_data_stock_market_sectors_2022.csv')
sectors_2022.head()
```

```python
df1 = df1.merge(sectors_2022[['Stock_code', 'Sector']], left_on='Code', right_on='Stock_code', how='left')
df1.drop('Stock_code', axis=1, inplace=True)
df1.head()
```

```python
df2 = pd.read_csv('../Data/NSE_data_all_stocks_2023.csv')
df2.head()  
```

```python
df3 = pd.read_csv('../Data/NSE_data_all_stocks_2024.csv')
df3.head()
```

```python
sectors_2023_2024 = pd.read_csv('../Data/NSE_data_stock_market_sectors_2023_2024.csv')
sectors_2023_2024.head()
```

```python
# to merge df2 and df3 with sectors_2023_2024
df2 = df2.merge(sectors_2023_2024[['Stock_code', 'Sector']], left_on='Code', right_on='Stock_code', how='left')
df2.drop('Stock_code', axis=1, inplace=True)
df3 = df3.merge(sectors_2023_2024[['Stock_code', 'Sector']], left_on='Code', right_on='Stock_code', how='left')
df3.drop('Stock_code', axis=1, inplace=True)
```

```python
df_columns = ['Date', 'Code', 'Name', '12m Low', '12m High', 'Day Low', 'Day High', 'Day Price', 'Previous', 'Change', '% Change', 'Volume', 'Adjusted Price', 'Sector']

df.columns = df_columns
df1.columns = df_columns
df2.columns = df_columns
df3.columns = df_columns
```

```python
# to merge all the dataframes into one dataframe
df_all = pd.concat([df, df1, df2, df3], ignore_index=True)
df_all.head()
```

```python
# first 10 where sector is null
df_all[df_all['Sector'].isna()].head(10)
```

```python
# % change where '-' is present
df_all[df_all['% Change'] == '-']
```

```python
cols = [
    "12m Low","12m High","Day Low","Day High",
    "Day Price","Previous","Change","% Change", "Volume","Adjusted Price"
]

# remove % first
df_all["% Change"] = df_all["% Change"].str.replace("%", "", regex=False)

# remove thousand separators + replace '-' + cast
df_all[cols] = (
    df_all[cols]
        .replace(",", "", regex=True)
        .replace("-", 0)
        .apply(pd.to_numeric, errors="coerce")
)
```

```python
df_all['Date'] = pd.to_datetime(df_all['Date'], errors='coerce')
```
