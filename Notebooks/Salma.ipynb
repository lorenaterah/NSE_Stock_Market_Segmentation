### Feature Engineering

Steps:

    1. Derive Daily Returns

    2. Derive Rolling Volatility 
    
    3. Derive Maximum drawdown 


```python
df = pd.read_csv('../Data/NSE_data_cleaned.csv')
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d", errors="coerce")
df["Date"] = df["Date"].dt.strftime("%Y-%m")
```

```python
df.dropna(inplace=True)
```

```python
df['Change'] = df['Day Price'] - df['Previous']
df['%Change'] = (df['Change']/df['Previous']) * 100
df
```

```python
df.loc[df["Adjusted Price"] == 0, "Adjusted Price"] = df.loc[df["Adjusted Price"] == 0, "Day Price"]
df.head()
```

```python
df = df.sort_values(["Code", "Date"])

df["return"] = (
    df.groupby("Code")["Adjusted Price"]
      .pct_change()
)

df["log_return"] = (
    np.log(df["Adjusted Price"])
      .groupby(df["Code"])
      .diff()
)
df
```

```python
df['return'].value_counts()
```

```python
df["daily_range"] = df["Day High"] - df["Day Low"]
df["range_pct"] = df["daily_range"] / df["Previous"]

df["high_low_ratio"] = df["Day High"] / df["Day Low"]
```

```python
df["ma_7"] = df.groupby("Code")["Day Price"].transform(lambda x: x.rolling(7).mean())
df["ma_14"] = df.groupby("Code")["Day Price"].transform(lambda x: x.rolling(14).mean())

df["momentum_7"] = df["Day Price"] / df["ma_7"] - 1
df["momentum_14"] = df["Day Price"] / df["ma_14"] - 1
```

```python
df["volatility_7"] = df.groupby("Code")["return"].transform(lambda x: x.rolling(7).std())
df["volatility_14"] = df.groupby("Code")["return"].transform(lambda x: x.rolling(14).std())
```

```python
df["vol_ma_7"] = df.groupby("Code")["Volume"].transform(lambda x: x.rolling(7).mean())
df["vol_ratio"] = df["Volume"] / df["vol_ma_7"]
df["position_12m"] = (df["Day Price"] - df["12m Low"]) / (df["12m High"] - df["12m Low"])
```

```python
df["target_up"] = (df.groupby("Code")["Day Price"]
                     .shift(-1) > df["Day Price"]).astype(int)
```

```python
df["target_return"] = df.groupby("Code")["return"].shift(-1)
df.columns
```

```python
features = ['return', 'log_return','daily_range', 'range_pct', 'momentum_7', 'momentum_14', 'volatility_7', 'volatility_14',
        'vol_ratio', 'position_12m']

df_model = df.copy()

df_model = df_model.sort_values(["Code","Date"])

# forward fill features within each stock
df_model[features] = df_model.groupby("Code")[features].ffill()

# drop rows where target is NaN
df_model = df_model.dropna(subset=["target_return"])
```

```python
# Assuming df is sorted by Stock_code and Date
df['running_max'] = df.groupby('Code')['Day Price'].cummax()
df['drawdown'] = (df['Day Price'] - df['running_max']) / df['running_max']
df.head()
```
